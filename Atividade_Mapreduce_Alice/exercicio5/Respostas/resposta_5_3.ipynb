O exercício 5.3 realiza uma análise de vendas comparando Python puro e PySpark em datasets de diferentes tamanhos. Primeiro, o código gera arquivos de vendas simulados com informações como produto, categoria, quantidade, preço, vendedor, região e timestamp. Em Python puro, ele calcula o valor total de cada transação, identifica os 10 produtos mais lucrativos e a receita por categoria, medindo o tempo de execução. Em PySpark, o mesmo cálculo é feito de forma distribuída, permitindo processar grandes volumes de dados rapidamente e testando o desempenho em diferentes números de cores de processamento. O código também compara os tempos de execução entre Python puro e Spark, mostrando o ganho de performance (speedup) e gera gráficos que ilustram como o Spark escala melhor à medida que o dataset aumenta, demonstrando sua eficiência e escalabilidade para análises de Big Data.
