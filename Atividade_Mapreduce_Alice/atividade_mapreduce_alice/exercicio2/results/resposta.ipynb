Resumo da Solução

1. Bandwidth Analysis:
MAP: Usa a URL como chave e o size (convertido para int) como valor.
REDUCE: Soma todos os valores de size agrupados por URL.

2. Error Rate:
MAP: Usa a hora como chave. O valor é uma tupla (is_error, 1), onde is_error é 1 se o status for 4xx/5xx e 0 caso contrário, e 1 representa o total de requisições.
REDUCE: Soma as tuplas (is_error, 1) para obter o total de erros e o total de requisições por hora. Em seguida, calcula a taxa de erros (total_erros / total_req).

Conclusão do Exercício de Análise de Logs de Servidor Web

O projeto de análise de logs de servidor web utilizando a abordagem MapReduce demonstrou ser altamente eficaz na extração de insights operacionais. As análises fundamentais confirmaram que a página /index.html é, o recurso mais acessado, e a maior parte das requisições resultou em um código de status 200 (OK), indicando a saúde básica. 

No entanto, os desafios adicionais forneceram uma visão mais profunda sobre o desempenho e a estabilidade. A Análise de Bandwidth revelou as URLs que consomem mais recursos de rede, permitindo a otimização de conteúdo. Mais criticamente, a Taxa de Erros por Hora identificou os períodos de pico de instabilidade (como o registro de 4xx e 5xx), apontando para a necessidade de investigar falhas na aplicação durante horários específicos, como as 09:00 e 10:00. Em suma, a aplicação do MapReduce não apenas extraiu estatísticas de volume, mas também forneceu métricas de desempenho.
