Conclusão de COM Combiner e SEM Combiner

Ao processar grandes volumes de dados em MapReduce, utilizar um combiner apresenta vantagens significativas sobre não utilizá-lo. Sem combiner, todos os pares chave-valor gerados pela fase de mapeamento são enviados diretamente ao reduce, o que aumenta consideravelmente o tempo de execução, consome mais memória e gera um volume muito maior de dados no shuffle. Já com combiner, ocorre uma mini-redução local, agregando os valores antes de enviá-los ao reduce, o que diminui drasticamente o número de pares transferidos, reduz o consumo de memória e acelera o processamento. Em datasets grandes, o uso do combiner pode tornar a execução de 2 a 5 vezes mais rápida, reduzir em até 90% o tráfego de dados no shuffle e diminuir significativamente o pico de memória necessário, tornando o processamento mais eficiente e escalável.

-Tempo de Execução:

Sem combiner:

Cada chave emitida pelo map vai diretamente para o reduce.
Todas as tuplas são transferidas e armazenadas em memória no shuffle.
O reduce precisa processar todos os valores, mesmo que várias tuplas possam ser pré-agregadas.

Com combiner:

O combiner faz uma pré-agregação local nos nós onde o map é executado.
Isso reduz o número de pares chave-valor que precisam ser enviados para o shuffle.
Resultado: menos dados para reduzir → tempo total de execução menor.

Resumo: Combiner geralmente diminui significativamente o tempo de execução, especialmente com grandes volumes de dados ou muitas tuplas repetidas por chave.

-Memória Utilizada:

Sem combiner:

Toda a lista de valores gerados pelo map precisa ser mantida em memória até o reduce.
Se houver muitas transações repetidas (mesma chave), isso aumenta o uso de memória.

Com combiner:

Valores duplicados ou que podem ser agregados localmente já são somados antes do shuffle.
Menos itens em memória → redução no uso de memória nos nós intermediários.

Resumo: Combiner ajuda a economizar memória, principalmente em clusters com muitas tuplas repetidas.

-Dados Transferidos no Shuffle

Sem combiner:

Todos os pares (chave, valor) são enviados do map para o reduce.
Se uma chave aparece 1000 vezes, todos os 1000 pares são enviados pelo shuffle.

Com combiner:

Pré-agregação reduz o número de pares. Por exemplo, 1000 transações do mesmo produto podem virar 1 par (produto, soma_valores).
Menos dados transferidos → rede e disco são menos sobrecarregados.

Resumo: Combiner reduz drasticamente os dados transferidos entre map e reduce, o que é crítico em clusters distribuídos grandes.
